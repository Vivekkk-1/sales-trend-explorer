```python
import pandas as pd

# Load the dataset
try:
    df = pd.read_csv("/content/sample_data/Sample.csv", encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv("/content/sample_data/Sample.csv", encoding='latin1')


# Show top 5 rows
df.head()
```

```python
# Check for missing values
df.isnull().sum()

# Check datatypes
df.dtypes

# Convert Order Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Create new columns
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
```

```python
monthly_sales = df.groupby(['Year', 'Month'])['Sales'].sum().reset_index()

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(monthly_sales['Sales'])
plt.title("Monthly Sales Trend")
plt.xlabel("Month Number")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5)
top_products.plot(kind='barh', title="Top 5 Selling Products", figsize=(10,5))
```

```python
region_sales = df.groupby('Region')['Sales'].sum()
region_sales.plot(kind='pie', autopct='%1.1f%%', title='Sales by Region', figsize=(6,6))
```
